```python
import numpy as np 
import pandas as pd                                                     
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("bank-additional-full.csv", sep=";")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['converted'] = df['y'].apply(lambda  c:1 if c== 'yes' else 0)
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt 

def plot_corr(df, size=10):
    
    corr = df.corr()
    fig, ax  = plt.subplots(figsize=(size,size))
    ax.legend()
    cax = ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation = "vertical")
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(df)
```

```python
print('Total converted: %i out of %i' % (df.converted.sum(),len(df)))
```

```python
print('Our conversion rate: %0.2f%%' % (df.converted.sum()/len(df) * 100.0))
```

```python
pd.DataFrame(df.groupby(by='campaign')['converted'].sum())
```

```python
conv_by_camp = df.groupby('campaign')['converted'].sum()/df.groupby('campaign')['converted'].count()*100.0
conv_by_camp
```

```python
age_con = df.groupby(['age','converted']).agg({'converted':'count'})
age_con2 = age_con.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
age_con2
```

```python
ax = age_con2.plot(grid=True, figsize=(10, 7), title=('conversion by age'))
ax.set_xlabel('Age')
ax.set_ylabel('Conversion rate (%)')
                  
plt.show()
```

```python
df['age_group'] = df['age'].apply(lambda x: '[17,25)' if x< 25 else '[25,35)'
                                  if x < 35 else '[35,45)'
                                  if x < 45 else '[45,55)'
                                  if x < 55 else '[55,65)'
                                  if x < 65 else '65+')

df.head()
```

```python
pd.DataFrame(df.groupby('age_group')['converted'].sum())
```

```python
pd.DataFrame(df.groupby('age_group')['converted'].count())
```

```python
conv_age_group = df.groupby('age_group')['converted'].sum()/df.groupby('age_group')['converted'].count() *100

conv_age_group
```

```python
ax = conv_age_group.loc[['[17,25)','[25,35)','[35,45)','[45,55)','[55,65)','65+' ]].plot(
    kind='bar',
    color='green',
    grid=True,
    figsize=(12, 8),
    title='Conversion rate by age group')

ax.set_xlabel('age')
ax.set_ylabel('conversion rate %')

plt.show()
```

```python
conv_edu = pd.pivot_table(df, values = 'y', index='education', columns = 'converted', aggfunc=len)
conv_edu
```

```python
conv_edu.plot(
    kind='pie',
    figsize=(15,7),
    startangle=90,
    subplots=True,
    autopct=lambda x : '%0.2f%%' % x,
    legend=False
    )
```

```python
new_order = [1,0]
conv_edu = conv_edu[conv_edu.columns[new_order]]
conv_edu
```

```python
ax = conv_edu.loc[['basic.4y', 'basic.6y', 'basic.9y', 'high.school', 'illiterate','professional.course','university.degree','unknown']].plot(
    kind='bar',
    grid=True,
    stacked=True,
    figsize=(10,7))

ax.set_title('conversions by education')
ax.set_xlabel('education level')
ax.set_ylabel('conversion counts')
    
plt.show()
```
